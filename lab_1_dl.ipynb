```python
# 1- libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# 2- Importing the dataset
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Lab1_dl/Salary_Data.csv')
```

```python
# 3- split data
X = dataset.iloc[:, :-1].values #independent variable array
y = dataset.iloc[:, 1].values #dependent variable vector
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3,
random_state = 0)

# 4- plot the x & y
plt.scatter(X,y,color='Green')
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
sc_y = StandardScaler()
y_train = sc_y.fit_transform(y_train.reshape(-1,1))
```

```python
# 5- build the model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# 6- Predicting the Test set results
y_pred = regressor.predict(X_test)
```

```python
# 7- Visualising the Test set results
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train , regressor.predict(X_train) , color = "Green")


plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# 8- Visualising the Training set results with different colors
plt.scatter(X_train, y_train, color = 'pink')
plt.plot(X_train, regressor.predict(X_train), color = 'violet')
```
